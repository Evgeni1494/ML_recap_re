```python
from pymongo import MongoClient
import pandas as pd
from dotenv import load_dotenv
import os
load_dotenv()
ATLAS_KEY= os.getenv('ATLAS_KEY')

client= MongoClient(ATLAS_KEY)

db=client.cluster0
collection=db["boxoff_us_1st"]
collection_fr=db["boxoff_FRA_1st"]
collection_frfr=db["nbre_entrees_10kgood"]
```

```python
data=collection.find()
data_fr=collection_fr.find()
data_frfr=collection_frfr.find()
df_us=pd.DataFrame(data)
df_fr=pd.DataFrame(data_fr)
df_frfr=pd.DataFrame(data_frfr)
```

```python
df_us['title'] = df_us['title'].str.lower().str.strip()
df_frfr['title'] = df_frfr['title'].str.lower().str.strip()
```

```python
df_merged = df_fr.merge(df_us, on='title', suffixes=('_fr', '_us'), how='left')
```

```python
df_merged_unique = df_merged.drop_duplicates(subset='title')
```
